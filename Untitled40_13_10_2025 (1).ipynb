## Suppression des lignes contenant ces outliers :

```python


print(f"Taille du dataset avant suppression des outliers : {data.shape}")
print("-" * 50)
data_sans_outliers = data.copy()

for column in data_sans_outliers.select_dtypes(include='number').columns:

    # 1. Calcul des bornes (méthode IQR)
    Q1 = data_sans_outliers[column].quantile(0.25)
    Q3 = data_sans_outliers[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 2. Identification des outliers
    outliers = data_sans_outliers[(data_sans_outliers[column] < lower_bound) | (data_sans_outliers[column] > upper_bound)]

    # 3. Affichage du nombre d'outliers trouvés
    if not outliers.empty:
        print(f"Nombre de valeurs aberrantes pour '{column}' : {len(outliers)}")
    data_sans_outliers = data_sans_outliers[(data_sans_outliers[column] >= lower_bound) & (data_sans_outliers[column] <= upper_bound)]

print("-" * 50)
print(f"Taille du dataset après suppression des outliers : {data_sans_outliers.shape}")
print(f"{data.shape[0] - data_sans_outliers.shape[0]} lignes ont été supprimées.")

# On remplace l'ancien dataframe par le nouveau, nettoyé des outliers
data = data_sans_outliers
```

## Standardisation et split de la base des donnees :

```python


# === 3. Définir les variables ===
X = data[['F(mL/mn)', 't(mn)']]
targets = ['CO2 selectivity (%)', 'H2/CO ratio', 'E reformage (kJ/min)']

# === 4. Standardisation ===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === 5. Split train/test ===
X_train, X_test, y_train_full, y_test_full = train_test_split(
    X_scaled, data[targets], test_size=0.3, random_state=42
)
```

## Construction du Model_2 pour la prediction de CO selectivity (%):

```python
optimizers = ['adam', 'rmsprop', 'adamax']
activations = ['relu', 'sigmoid', 'tanh']

# === Fonction pour construire le modèle ===
def build_model_2(n, act, optimizer_name):
    if optimizer_name == 'adam':
        opt = Adam(0.01)
    elif optimizer_name == 'rmsprop':
        opt = RMSprop(0.01)
    elif optimizer_name == 'adamax':
        opt = Adamax(0.01)
    else:
        raise ValueError("Optimizer inconnu.")

    model = Sequential([
        Input(shape=(2,)),
        Dense(n, activation=act),
        Dense(1)
    ])
    model.compile(optimizer=opt, loss='mse')
    return model
```

```python

results = []
# === Variables pour suivre le meilleur modèle ===
best_mse = float('inf') # On commence avec un MSE infini pour que le premier soit forcément meilleur
model_opt = None        # Variable qui stockera le meilleur modèle
best_params = {}
for opt in optimizers:
    for act in activations:
        print(f"\n===== Optimizer: {opt.upper()} | Activation: {act.upper()} =====")

        for n in range(2, 20):
            # === Construction du modèle ===
            model_2 = build_model_2(n, act, opt)

            # === Données cibles ===
            Y2_train = y_train_full['CO2 selectivity (%)']
            Y2_test = y_test_full['CO2 selectivity (%)']

            # === Entraînement ===
            history = model_2.fit(
                X_train, Y2_train,
                epochs=150,
                batch_size=8,
                verbose=0,
                validation_split=0.2
            )

            y_pred = model_2.predict(X_test).flatten()

            r2 = r2_score(Y2_test, y_pred)
            rmse = np.sqrt(mean_squared_error(Y2_test, y_pred))
            mape = mean_absolute_percentage_error(Y2_test, y_pred)
            mse = mean_squared_error(Y2_test, y_pred)

            print(f"n={n:2d} | R²={r2:.3f} | RMSE={rmse:.3f} | MAPE={mape:.3f} | MSE={mse:.3f}")

            results.append({
                'optimizer': opt,
                'activation': act,
                'n': n,
                'r2': r2,
                'rmse': rmse,
                'mape': mape,
                'mse': mse
            })

results_df_SCO= pd.DataFrame(results)
for opt in optimizers:
    plt.figure(figsize=(10, 5))
    subset_opt = results_df_SCO[results_df_SCO['optimizer'] == opt]

    for act in activations:
        subset = subset_opt[subset_opt['activation'] == act]
        plt.plot(subset['n'], subset['mse'], marker='o', label=f"{act}")

    plt.title(f"Évolution du MSE selon le nombre de neurones — Optimizer : {opt.upper()}")
    plt.xlabel("Nombre de neurones (n)")
    plt.ylabel("MSE (Mean Squared Error)")
    plt.legend(title="Activation")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

```
